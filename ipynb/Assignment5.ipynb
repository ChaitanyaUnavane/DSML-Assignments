```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("/content/car_crashes.csv")
```

```python
#Scatterplot
plt.scatter(df.speeding,df.total)
plt.xlabel("Speeding")
plt.ylabel("Total")
plt.title("Scatterplot")
sns.set_style("darkgrid", {'grid.color': '.75'})
plt.show()
```

```python
#Barplot
plt.bar(df.speeding , df.not_distracted, width=0.2)
plt.xlabel("Speeding")
plt.ylabel("Not Distracted")
plt.title("Barplot")
plt.show()
```

```python
#Histogram
plt.hist(df.speeding)
plt.xlabel("Speeding")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
#Regplot
sns.regplot(x=df.alcohol, y=df.speeding)
plt.title("Regeression Plot")
plt.show()
```

```python
#Pointplot
plt.plot(df['total'], marker="o", linestyle="-")
plt.xlabel("Data Points")
plt.ylabel("Total Crashes")
plt.title("Pointplot")
plt.show()
```
